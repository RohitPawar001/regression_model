```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import pickle
```

# independent and dependent feature split

```python
# independent features
x = df.drop(['AveBedrms', 'Population', 'AveOccup','Latitude', 'Longitude','Price'],axis=1)
```

```python
# dependent feature
y = df["Price"]
```

# train and test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# standardizing the features

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# regression model

```python
reg_model = LinearRegression()
```

```python
reg_model.fit(x_train,y_train)
```

# coefficients and intercept s of the model

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
pred = reg_model.predict(x_test)
```

# scatterplot of predictions

```python
plt.scatter(y_test,pred)
plt.xlabel("actual_values")
plt.ylabel("predicted_values")
```

# residuals


```python
residuals = y_test-pred
```

```python
sns.displot(residuals,kind="kde")
```

# model evaluation

```python
print("mse :-",mean_squared_error(y_test,pred))
print("r2_score :-",r2_score(y_test,pred))
```

```python
reg_model.predict(scaler.transform(np.array(x.iloc[0]).reshape(1,-1)))
```
